# 💬 Social & Inbox

Two complementary sub-clients:
- **InboxClient** — signed REST API for messaging and notifications
- **SocialClient** — ActivityPub-based likes and follows

Both use EdDSA request signing from the Keypairoom keypair.

## Setup

> ⚠️ Both clients require authentication.

```javascript
(async () => {
  (async () => {
    const { InterfacerClient } = await import('https://esm.sh/@dyne/interfacer-client');
    const { SANDBOX_CONFIG } = await import('./config.js');
    
    const client = new InterfacerClient(SANDBOX_CONFIG);
    
    console.log('🔐 Authenticated:', client.isAuthenticated());
    console.log('📧 Inbox send URL:', client.config.inbox?.send || '(not configured)');
    console.log('👤 Social person base:', client.config.social?.personBase || '(not configured)');
    
    if (!client.isAuthenticated()) {
      console.log('\n⚠️ Not authenticated! Run 01_Authentication_and_Crypto.ipynb first.');
    }    
    // Expose to other cells via window
    if (!window._if) window._if = {};
    window._if.InterfacerClient = InterfacerClient;
    window._if.SANDBOX_CONFIG = SANDBOX_CONFIG;
    window._if.client = client;
  })();
  
  // Expose to other cells via window
  if (!window._if) window._if = {};
  window._if.InterfacerClient = InterfacerClient;
  window._if.SANDBOX_CONFIG = SANDBOX_CONFIG;
  window._if.client = client;
})();
```

## 📧 Inbox: Send a Message

Messages are sent to one or more receivers with a subject and payload.

```javascript
(async () => {
  // Get shared state
  const _if = window._if || {};
  const InterfacerClient = _if.InterfacerClient;
  const SANDBOX_CONFIG = _if.SANDBOX_CONFIG;
  const client = _if.client;
  (async () => {
    // Get shared state
    const _if = window._if || {};
    const InterfacerClient = _if.InterfacerClient;
    const SANDBOX_CONFIG = _if.SANDBOX_CONFIG;
    const client = _if.client;
    const userId = client.store.getItem('authId');
    
    if (userId) {
      try {
        await client.inbox.sendMessage(
          { proposalId: 'example-proposal-123', note: 'Hey! Check out my contribution.' },
          [userId],  // Send to self for demo
          'New Contribution Proposal'
        );
        console.log('✅ Message sent!');
        console.log('  From:', userId);
        console.log('  Subject: New Contribution Proposal');
      } catch (err) {
        console.warn('⚠️ Send failed (sandbox inbox may not be available):', err.message);
      }
    } else {
      console.log('⚠️ Not authenticated.');
    }
  })();
})();
```

## 📥 Inbox: Read Messages

```javascript
(async () => {
  // Get shared state
  const _if = window._if || {};
  const InterfacerClient = _if.InterfacerClient;
  const SANDBOX_CONFIG = _if.SANDBOX_CONFIG;
  const client = _if.client;
  (async () => {
    // Get shared state
    const _if = window._if || {};
    const InterfacerClient = _if.InterfacerClient;
    const SANDBOX_CONFIG = _if.SANDBOX_CONFIG;
    const client = _if.client;
    try {
      const messages = await client.inbox.getMessages();
      console.log('📬 Messages:', messages.length);
      
      if (messages.length > 0) {
        const msg = messages[0];
        console.log('\n📋 Latest message:');
        console.log('  ID:', msg.id);
        console.log('  From:', msg.sender);
        console.log('  Read:', msg.read);
        console.log('  Date:', msg.content.data);
        console.log('  Subject:', msg.content.subject);
        console.log('  Body:', JSON.stringify(msg.content.message).substring(0, 100));
      }
      
      // Only unread
      const unreadOnly = await client.inbox.getMessages(true);
      console.log('🔴 Unread messages:', unreadOnly.length);
    } catch (err) {
      console.warn('⚠️ Read messages failed:', err.message);
    }
  })();
})();
```

## ❤️ Social: Like a Resource

The Social client uses **ActivityPub** (ActivityStreams 2.0) protocol.
Likes and Follows are sent to the user's outbox.

```javascript
(async () => {
  // Get shared state
  const _if = window._if || {};
  const InterfacerClient = _if.InterfacerClient;
  const SANDBOX_CONFIG = _if.SANDBOX_CONFIG;
  const client = _if.client;
  (async () => {
    // Get shared state
    const _if = window._if || {};
    const InterfacerClient = _if.InterfacerClient;
    const SANDBOX_CONFIG = _if.SANDBOX_CONFIG;
    const client = _if.client;
    const sampleResourceId = 'example-resource-id';
    
    try {
      await client.social.likeResource(sampleResourceId);
      console.log('❤️  Liked resource:', sampleResourceId);
      console.log('   Activity sent to outbox as ActivityStreams Like');
    } catch (err) {
      console.warn('⚠️ Like failed (may need a real resource):', err.message);
    }
    
    // The underlying ActivityStreams structure:
    console.log('\n📋 Like activity structure:');
    console.log('  @context: https://www.w3.org/ns/activitystreams');
    console.log('  type: Like');
    console.log('  actor: <personBase>/<userId>');
    console.log('  object: <economicResourceBase>/<resourceId>');
    console.log('  published: ISO 8601 timestamp');
  })();
})();
```

## 🔍 Social: Query Likes & Followers

```javascript
(async () => {
  // Get shared state
  const _if = window._if || {};
  const InterfacerClient = _if.InterfacerClient;
  const SANDBOX_CONFIG = _if.SANDBOX_CONFIG;
  const client = _if.client;
  (async () => {
    // Get shared state
    const _if = window._if || {};
    const InterfacerClient = _if.InterfacerClient;
    const SANDBOX_CONFIG = _if.SANDBOX_CONFIG;
    const client = _if.client;
    try {
      // Get all resources the user has liked
      const likes = await client.social.getLikes();
      console.log('❤️  Liked resources:', likes.length, 'items');
      
      // Check if a specific resource is liked
      const isLiked = await client.social.isLiked(sampleResourceId);
      console.log('   Is liked?', isLiked);
    } catch (err) {
      console.warn('⚠️ Query likes failed:', err.message);
    }
    
    try {
      // Get followers of a resource
      const followers = await client.social.getFollowers(sampleResourceId);
      console.log('👥 Followers:', followers.length, 'users');
      
      // Get resources the user is following
      const following = await client.social.getFollowing();
      console.log('🔔 Following:', following.length, 'resources');
      
      // Check if following a resource
      const isFollowing = await client.social.isFollowing(sampleResourceId);
      console.log('   Is following?', isFollowing);
    } catch (err) {
      console.warn('⚠️ Query followers failed:', err.message);
    }
  })();
})();
```

## 📋 API Reference

### InboxClient

| Method | Description |
|---|---|
| `sendMessage(message, receivers, subject?)` | Send to one or more users |
| `getMessages(onlyUnread?)` | Read inbox, sorted by date desc |
| `getUnreadCount()` | Count unread messages |
| `markRead(messageId)` | Mark message as read |

### SocialClient

| Method | Description |
|---|---|
| `likeResource(resourceId)` | Like via ActivityPub Like |
| `followResource(resourceId)` | Follow via ActivityPub Follow |
| `getLikes()` | Get liked resource IDs |
| `isLiked(resourceId)` | Check if resource is liked |
| `getFollowers(resourceId)` | Get list of follower IDs |
| `getFollowing()` | Get followed resource IDs |
| `isFollowing(resourceId)` | Check if following resource |
