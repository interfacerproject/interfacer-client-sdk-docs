```javascript
(async () => {
  // Test: import from /files/ (absolute path)
  try {
    const mod = await import('/files/config.js');
    console.log('Absolute /files/config.js: OK ->', Object.keys(mod));
  } catch(e) {
    console.error('Absolute /files/config.js: FAIL ->', e.message);
  }
})();
```

```javascript
(async () => {
  // Test: import from ./files/ (relative to origin)
  try {
    const mod = await import(new URL('/files/config.js', location.origin).href);
    console.log('URL-based /files/config.js: OK ->', Object.keys(mod));
  } catch(e) {
    console.error('URL-based /files/config.js: FAIL ->', e.message);
  }
})();
```

```javascript
(async () => {
  // Test: relative from kernel ext to files (4 dirs up)
  try {
    const mod = await import('../../../../files/config.js');
    console.log('Relative ../../../../files/config.js: OK ->', Object.keys(mod));
  } catch(e) {
    console.error('Relative ../../../../files/config.js: FAIL ->', e.message);
  }
})();
```
