## 🏗️ Architecture: The Facade Pattern

The SDK is organized as a **Facade** — a single entry point (`InterfacerClient`) that provides lazy access to specialized sub-clients.

```mermaid
graph TD
    Client[InterfacerClient]
    Config[InterfacerConfig]
    Store[KeyStorage]
    GraphQL[GraphQLClient]
    
    Client --> Config
    Client --> Store
    Client --> Auth[auth — Keypairoom, EdDSA]
    Client --> Resources[resources — Projects, Machines]
    Client --> Files[files — SHA-512, SHA-256]
    Client --> DPP[dpp — Digital Product Passport]
    Client --> Inbox[inbox — Messaging]
    Client --> Wallet[wallet — Idea/Strength points]
    Client --> Social[social — ActivityPub likes/follows]
    Client --> Tagging[tagging — Classification]
    Client --> Import[import — GitHub/GitLab]
    
    Auth --> GraphQL
    Resources --> GraphQL
    GraphQL --> Store
```

**All sub-clients share the same config and key storage.**
They are initialized lazily — you only pay for what you use.

## 2. Configuration

The `InterfacerConfig` object tells the SDK where to find each backend service.

### Option A: `proxyUrl` (simplest)

A single `proxyUrl` auto-derives all other URLs using well-known paths:

```js
{ proxyUrl: 'https://proxy.dpp-dev.ddns.dyne.org' }
```

### Option B: Explicit URLs (full control)

```js
{
  zenflowsUrl: '.../zenflows/api',
  dppUrl: '.../interfacer-dpp',
  inbox: { send: '...', read: '...', countUnread: '...', setRead: '...' },
  walletUrl: '.../wallet/token',
  social: { personBase: '...', economicResourceBase: '...' },
}
```

````javascript
(async () => {
  // Get shared state
  const _if = window._if || {};
  const InterfacerClient = _if.InterfacerClient;
  const SANDBOX_CONFIG = _if.SANDBOX_CONFIG;
  const client = _if.client;
  const createConfig = _if.createConfig;
  const deriveEndpointsFromProxy = _if.deriveEndpointsFromProxy;
  (async () => {
    // Get shared state
    const _if = window._if || {};
    const InterfacerClient = _if.InterfacerClient;
    const SANDBOX_CONFIG = _if.SANDBOX_CONFIG;
    const client = _if.client;
    const createConfig = _if.createConfig;
    const deriveEndpointsFromProxy = _if.deriveEndpointsFromProxy;
    // See all derived endpoints from the proxy URL
    const endpoints = deriveEndpointsFromProxy('https://proxy.dpp-dev.ddns.dyne.org');
    
    console.log('📋 Derived endpoints:');
    console.log('  Zenflows API:', endpoints.zenflowsUrl);
    console.log('  DPP:', endpoints.dppUrl);
    console.log('  Inbox send:', endpoints.inbox.send);
    console.log('  Wallet:', endpoints.walletUrl);
    console.log('  Social person:', endpoints.social.personBase);
    console.log('  OSH:', endpoints.oshUrl);
  })();
})();
````

## 5. Key Persistence (localStorage)

The SDK persists cryptographic keys and auth state in `localStorage` by default.
This means:

- **Between notebooks:** Auth state survives browser refreshes
- **Security warning:** localStorage is not encrypted — use `createMemoryStorage()` for sensitive contexts
- **Custom storage:** Pass any `KeyStorage` implementation to the constructor

````javascript
(async () => {
  // Get shared state
  const _if = window._if || {};
  const InterfacerClient = _if.InterfacerClient;
  const SANDBOX_CONFIG = _if.SANDBOX_CONFIG;
  const client = _if.client;
  const createConfig = _if.createConfig;
  const deriveEndpointsFromProxy = _if.deriveEndpointsFromProxy;
  (async () => {
    // Get shared state
    const _if = window._if || {};
    const InterfacerClient = _if.InterfacerClient;
    const SANDBOX_CONFIG = _if.SANDBOX_CONFIG;
    const client = _if.client;
    const createConfig = _if.createConfig;
    const deriveEndpointsFromProxy = _if.deriveEndpointsFromProxy;
    // Show current localStorage keys managed by the SDK
    const knownKeys = [
      'eddsaPrivateKey', 'eddsaPublicKey',
      'ethereumPrivateKey', 'ethereumAddress',
      'reflowPrivateKey', 'reflowPublicKey',
      'bitcoinPrivateKey', 'bitcoinPublicKey',
      'ecdhPrivateKey', 'ecdhPublicKey',
      'seed', 'authId', 'authName', 'authEmail', 'authUsername'
    ];
    
    console.log('💾 localStorage state:');
    for (const key of knownKeys) {
      const val = client.store.getItem(key);
      if (val) {
        const display = key.includes('Private') || key === 'seed'
          ? val.substring(0, 10) + '...'
          : val;
        console.log(`  ${key}: ${display}`);
      }
    }
    
    // Custom storage example (in-memory, non-persistent)
    const { createMemoryStorage } = await import('https://esm.sh/@dyne/interfacer-client');
    const ephemeralStore = createMemoryStorage();
    const ephemeralClient = new InterfacerClient(SANDBOX_CONFIG, ephemeralStore);
    console.log('\n🧪 Ephemeral client (no persistence):', !!ephemeralClient);
  })();
})();
````
